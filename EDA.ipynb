# Exploratory Data Analysis (EDA)
Bank Customer Churn Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Customer-Churn-Records.csv')
df.head()
```

```python
df.shape
df.info()
```

```python
df['Exited'].value_counts(normalize=True)
```

```python
sns.countplot(x='Exited', data=df)
plt.show()
```

```python
sns.boxplot(x='Exited', y='Balance', data=df)
plt.show()

sns.boxplot(x='Exited', y='Age', data=df)
plt.show()
```

- A higher proportion of customers did not churn.
- Customers who churned tend to have higher balances on average.
- Age appears to differ between churners and non-churners. Older customers are more likely to churn.
